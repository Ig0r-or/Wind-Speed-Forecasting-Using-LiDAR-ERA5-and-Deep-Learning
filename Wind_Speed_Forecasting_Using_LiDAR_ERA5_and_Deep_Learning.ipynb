```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Carregar os dados observacionais
df_lidar = pd.read_csv('data/dataset.csv')

# 2. Criar a coluna de tempo (combinando ano, mês, dia, hora e minuto)
df_lidar['valid_time'] = pd.to_datetime(df_lidar[['year', 'month', 'day', 'hour', 'minute']])
df_lidar = df_lidar.set_index('valid_time')

# 3. Visualização da série temporal ws100 (100 metros)
plt.figure(figsize=(12, 5))
df_lidar['ws100'].plot(color='green', label='LiDAR ws100 (Obs)')
plt.title("Série Temporal LiDAR - Velocidade do Vento a 100m")
plt.ylabel("Velocidade (m/s)")
plt.legend()
plt.show()

# 4. Ver as primeiras linhas para conferir a estrutura
print(df_lidar[['ws100']].head())
```

```python
import xarray as xr
import matplotlib.pyplot as plt

# 1. Carregar o arquivo NetCDF (preguiça de mudar o nome do arquivo)

ds = xr.open_dataset('data/era5_dataset.nc')

# 2. Visualizar a estrutura do "cubo de dados", duas estruturas de visualização: a estrutura geral do dataset e um mapa específico da variável de vento u100 para o primeiro tempo válido.
print(ds)
ds.u100.isel(valid_time=0).plot()
plt.title("Mapa de Vento ERA5 - Área de Extração")
plt.show()

# 3. Ver apenas as variáveis disponíveis

print("\nVariáveis extraídas:", list(ds.data_vars))
```
