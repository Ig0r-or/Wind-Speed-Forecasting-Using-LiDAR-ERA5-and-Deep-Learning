```python
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 1. Carregar o conjunto de dados ERA5
# Usarei xr.open_dataset para ler o arquivo NetCDF
data_path = '/content/data/era5_dataset.nc'
ds = xr.open_dataset(data_path)

# 2. Selecionei as variáveis meteorológicas principais para o resumo
colunas_interesse = ['d2m', 't2m', 'sp', 'u100', 'v100', 'z']
# Filtrei apenas as colunas que existem no Dataset para evitar novos erros
colunas_presentes = [c for c in colunas_interesse if c in ds.data_vars]
tabela_estatistica = ds[colunas_presentes].to_dataframe().describe() # Convert to DataFrame for .describe()

print("Estatística Descritiva do ERA5:")
display(tabela_estatistica)

# 3. Verifiquei se existem dados nulos
print("\nVerificação de lacunas (dados faltantes):")
print(ds[colunas_presentes].isnull().sum())
```

```python
# Verifiquei o alcance temporal dos metadados (seu inicio e fim)
print(f"Início da série ERA5: {ds.valid_time.min().values}")
print(f"Fim da série ERA5:    {ds.valid_time.max().values}")

# Verifiquei a resolução temporal (deve ser de 1 hora)
diff = ds.valid_time.diff('valid_time').values
print(f"Frequência dos dados: {np.unique(diff)}")
```

```python
# 2. Visualização a estrutura do "cubo de dados", duas estruturas de visualização:
#a estrutura geral do dataset e um mapa específico da variável de vento u100 para o primeiro tempo válido.

ds.u100.isel(valid_time=0).plot()
plt.title("Mapa de Vento ERA5 - Área de Extração")
plt.show()
```

```python
# Criação do ws100 diretamente no Dataset(cubo)
ds['ws100'] = (ds.u100**2 + ds.v100**2)**0.5
ds.ws100.attrs['units'] = 'm/s'
ds.ws100.attrs['long_name'] = 'Velocidade do vento em 100m'

# Plotar o primeiro passo de tempo de toda a grade
ds.ws100.isel(valid_time=0).plot()
plt.title("Velocidade do Vento - Grade ERA5")
plt.show()
```

```python
# Visualização 1: Série Temporal Completa
plt.figure(figsize=(12, 4))

# Defino o período de interesse, no caso o partiodo de 15 de setembro até 08 de novembro
start_time = '2021-09-15T00:00:00'
end_time = '2021-11-08T23:00:00'

# Seleciono o subconjunto de dados para o período especifico (ANALISAR DEPOIS)
ds_subset = ds.sel(valid_time=slice(start_time, end_time))

# Calcular a média de ws100 sobre as dimensões espaciais (latitude e longitude)
# para obter uma série temporal 1D para o subconjunto.
ws100_mean_spatial = ds_subset['ws100'].mean(dim=['latitude', 'longitude'])

plt.plot(ds_subset['valid_time'], ws100_mean_spatial, label='ERA5 ws100 (Média Espacial)')
plt.title(f"Série Temporal da Velocidade do Vento (Média Espacial ERA5) de {start_time[:10]} a {end_time[:10]}")
plt.ylabel("m/s")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 1. Fazendo a comparação com as observações LiDAR de 10 em 10 minutos
# Lendo os dados .csv do LiDAR
df_lidar = pd.read_csv('/content/data/dataset.csv')

# Garantir que a coluna de tempo seja datetime
df_lidar['valid_time'] = pd.to_datetime(df_lidar[['year', 'month', 'day', 'hour', 'minute']])
df_lidar = df_lidar.set_index('valid_time')

# Selecionar o período de interesse no LiDAR
lidar_periodo = df_lidar['ws100'].loc['2021-09-15':'2021-11-08']

# Adicionar o cálculo de ws100 ao dataset 'ds' para garantir sua presença
ds['ws100'] = (ds.u100**2 + ds.v100**2)**0.5
ds.ws100.attrs['units'] = 'm/s'
ds.ws100.attrs['long_name'] = 'Velocidade do vento em 100m'

# 2. Preparar os dados ERA5 (Xarray) no ponto do site
# Selecionando o pixel e o recorte temporal (usando coordenadas aproximadas)
era5_ponto = ds.sel(latitude=-2.716, longitude=-42.566, method='nearest')
era5_periodo = era5_ponto.ws100.sel(valid_time=slice('2021-09-15', '2021-11-08'))

# 3. Plotagem Comparativa
plt.figure(figsize=(15, 6))
plt.plot(lidar_periodo.index, lidar_periodo.values, label='LiDAR (Observado - 10 min)', alpha=0.6, color='green')
plt.plot(era5_periodo.valid_time, era5_periodo.values, label='ERA5 (Reanálise - 1h)', marker='o', markersize=3, color='blue', linestyle='-')

plt.title("Comparação de Velocidade do Vento a 100m: LiDAR vs ERA5 (Set-Nov 2021)")
plt.xlabel("Data")
plt.ylabel("Velocidade do Vento (m/s)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
